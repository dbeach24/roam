# Equations for t-SNE

## Feature Space

$$ \mathbf{X}_{N \times D} = \big[ \mathbf{x}_1 \ldots \mathbf{x}_N \big]^T $$

## Metric Space

$ d_{ij} = d(x_i, x_j) $

1. $ d_{ij} \ge 0 $ (non-negativity)
2. $ d_{ij} = 0 \implies x_i = x_j $ (identity)
3. $ d_{ij} = d_{ji} $ (symmetry)
4. $ d_{ik} \le d_{ij} + d_{jk} $ (triangle inequality)

## Sparse Neighborhood Graph

$$ \mathcal{N}(i) = \{ j_1 \ldots j_k \} $$

### t-SNE Neighborhood to Probability

$$ p_{i|j} = { 
    \exp \big(- d_{ij}^2 / 2\sigma_i^2 \big)
    \over
    \ \sum_{k \ne i} \exp \big( - d_{ik}^2 / 2\sigma_i^2 \big)
    } $$
    
$$ p_{ij} = { p_{j|i} + p_{i|j} \over 2N } $$

## Sparse Connectivity Graph

$$ 0 \le w_{ij} $$

### Connectivity to Probability

$$ p_{i|j} = { w_{ij} \over \sum_{k \ne i} w_{ik} } $$

This formulation avoids overly distorted values due to outliers:

$$ p_{ij} = { p_{i|j} + p_{j|i} \over 2N }$$


## Sparse Probability Graph

$$ 0 \le P_{j|i} \le 1 $$

$$ \sum_{j \ne i} p_{j | i} = 1 $$

$$ p_{ij} = { p_{j|i} + p_{i|j} \over 2N } $$

$$ Q = \sum_{k \ne l} q_{kl}^{*} $$
